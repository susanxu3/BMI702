# TxGNN Custom Split Baseline

This notebook evaluates the TxGNN disease-node baseline on the custom split in `BMI702/split` and reports:

- `MRR`
- `R@1`, `R@5`, `R@10`, `R@50`
- `AUROC (per-disease mean)`
- `AUPRC (balanced 1:1)`
- `AUROC (balanced 1:1)`

The split CSVs use PrimeKG `node_index` values, so the helper script first maps those through `TxGNN/data/node.csv` and then into the TxGNN embedding index space.

```python
import sys
from pathlib import Path

PROJECT_DIR = Path('/content/drive/MyDrive/BMI Year1/BMI702/project/BMI702')
if not PROJECT_DIR.exists():
    raise FileNotFoundError(PROJECT_DIR)

if str(PROJECT_DIR) not in sys.path:
    sys.path.insert(0, str(PROJECT_DIR))

print('Project dir:', PROJECT_DIR)
```

```python
from custom_txgnn_baseline_eval import run_txgnn_custom_split_baseline

summary_df, per_disease_df, metadata = run_txgnn_custom_split_baseline(
    project_dir=PROJECT_DIR,
    split_dir=PROJECT_DIR / 'split',
    results_dir=PROJECT_DIR / 'split' / 'txgnn_baseline_results',
    balanced_repeats=100,
    seed=42,
)

display(summary_df.style.format({'TxGNN Baseline': '{:.4f}'}))
print(metadata)

print('\nSaved files:')
print('  ', metadata['results_dir'] + '/txgnn_baseline_summary.csv')
print('  ', metadata['results_dir'] + '/txgnn_baseline_per_disease.csv')
```
