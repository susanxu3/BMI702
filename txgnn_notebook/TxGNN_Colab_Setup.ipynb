# TxGNN Colab Setup

This notebook installs and imports `TxGNN` in a Google Colab runtime using the repository already stored in Google Drive.

- Runtime: Colab Python 3
- Recommended hardware: GPU
- Repository path: `/content/drive/MyDrive/BMI Year1/BMI702/project/BMI702/TxGNN`
- Install method: mount Google Drive and install the local repo in editable mode

## What This Does

1. Mounts Google Drive.
2. Uses the existing `TxGNN` repo from Drive instead of cloning a new copy.
3. Pins `numpy`, `scipy`, and `torchdata` to avoid Colab/DGL compatibility issues.
4. Installs `dgl` from DGL's torch-matched wheel index.
5. Installs your Drive-based `TxGNN` checkout with `%pip install -e`.
6. Restarts the runtime.
7. Verifies that `txgnn` imports cleanly from the Drive path.

```python
from google.colab import drive

drive.mount('/content/drive')
```

```python
import subprocess
import sys
from pathlib import Path

repo_dir = Path('/content/drive/MyDrive/BMI Year1/BMI702/project/BMI702/TxGNN')

if not repo_dir.exists():
    raise FileNotFoundError(f'TxGNN repo not found: {repo_dir}')

def pip_install(*packages):
    cmd = [sys.executable, '-m', 'pip', 'install', '-q', *packages]
    print('Running:', ' '.join(cmd))
    subprocess.check_call(cmd)

pip_install('--no-cache-dir', '--force-reinstall', 'numpy==1.26.4', 'scipy==1.13.1')

import torch

major, minor = (int(part) for part in torch.__version__.split('+')[0].split('.')[:2])
print(f'Detected torch version: {torch.__version__}')
dgl_repo = f'https://data.dgl.ai/wheels/torch-{major}.{minor}/repo.html'
print(f'Using DGL wheel index: {dgl_repo}')

if (major, minor) >= (2, 3):
    torchdata_spec = 'torchdata==0.8.0'
elif (major, minor) >= (2, 2):
    torchdata_spec = 'torchdata==0.7.1'
else:
    torchdata_spec = 'torchdata==0.6.0'

pip_install('--no-deps', torchdata_spec)
pip_install('--no-cache-dir', '-f', dgl_repo, 'dgl')
pip_install(
    '--no-cache-dir',
    'pandas',
    'scikit-learn',
    'matplotlib',
    'tqdm',
    'goatools',
    'networkx',
    'requests',
)

pip_install('-e', str(repo_dir))

print('TxGNN installation complete from Google Drive.')
print('Restart the Colab runtime now, then run the verification cells below.')
```
